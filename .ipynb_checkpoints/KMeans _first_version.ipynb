```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def preprocess_image(image):
    """Apply CLAHE to enhance local contrast."""
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    return clahe.apply(image)

def enhance_image_kmeans(image, n_clusters=8):
    """Segment intensities via KMeans. Return raw and normalized segmented images."""
    # Flatten image
    pixel_values = image.reshape(-1, 1).astype(np.float32)
    
    # Apply KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(pixel_values)
    centers = kmeans.cluster_centers_.astype(np.float32)
    labels  = kmeans.labels_

    # Map to cluster centers
    segmented_raw = centers[labels].reshape(image.shape)

    # Normalize to 0–255
    segmented_norm = cv2.normalize(segmented_raw, None, 0, 255, cv2.NORM_MINMAX)

    return segmented_raw.astype(np.uint8), segmented_norm.astype(np.uint8)

def show_results(orig, clahe, raw_kmeans, enhanced):
    """Display four images side by side with titles."""
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    titles = ['Original', 'CLAHE', 'Raw KMeans', 'Enhanced']
    imgs   = [orig, clahe, raw_kmeans, enhanced]
    for ax, img, title in zip(axes, imgs, titles):
        ax.imshow(img, cmap='gray', vmin=0, vmax=255)
        ax.set_title(title)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Replace with your actual image path
    img_path = '/Users/abubakr/Downloads/Spinal Cord Wedge Shape Plain.jpg'
    orig = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if orig is None:
        raise FileNotFoundError(f"Could not load image at {img_path}")

    clahe_img = preprocess_image(orig)
    raw_kmeans, enhanced_img = enhance_image_kmeans(clahe_img, n_clusters=8)

    show_results(orig, clahe_img, raw_kmeans, enhanced_img)
```
