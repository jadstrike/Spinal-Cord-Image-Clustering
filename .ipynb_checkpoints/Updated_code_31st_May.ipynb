```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score, adjusted_rand_score, accuracy_score
from scipy.optimize import linear_sum_assignment

def preprocess_image(image: np.ndarray) -> np.ndarray:
    """
    Apply CLAHE to enhance local contrast in a grayscale image.
    """
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    return clahe.apply(image)

def enhance_image_kmeans(image: np.ndarray, n_clusters: int = 8):
    """
    Segment intensities via KMeans on the input 2D (grayscale) image.
    Returns:
      - labels: (H*W,) array of cluster assignments per pixel
      - centers: (n_clusters,) array of cluster-center intensities (float32)
      - segmented_raw: (H, W) image reconstructed by mapping each pixel to its center (uint8)
      - segmented_norm: (H, W) normalized segmented image in [0,255] for display (uint8)
    """
    # Flatten image to shape (num_pixels, 1)
    pixel_values = image.reshape(-1, 1).astype(np.float32)

    # Apply KMeans clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(pixel_values)
    centers = kmeans.cluster_centers_.flatten().astype(np.float32)  # shape (n_clusters,)
    labels  = kmeans.labels_                                       # shape (num_pixels,)

    # Reconstruct segmented image (raw intensities = exact center values)
    segmented_raw = centers[labels].reshape(image.shape)

    # Normalize the raw segmented image for display (stretch to [0,255])
    segmented_norm = cv2.normalize(segmented_raw, None, 0, 255, cv2.NORM_MINMAX)

    return labels, centers, segmented_raw.astype(np.uint8), segmented_norm.astype(np.uint8)

def compute_segmentation_metrics(
    gt_mask: np.ndarray,
    pred_labels: np.ndarray,
    n_clusters: int
):
    """
    Compute Accuracy, Normalized Mutual Information (NMI), and Adjusted Rand Index (ARI)
    between a ground-truth mask and KMeans cluster labels.
    
    Args:
      - gt_mask: 2D array of integer labels (shape: H×W). Each distinct integer is a separate label.
      - pred_labels: 1D array of cluster assignments (length H*W), integers in [0..n_clusters-1].
      - n_clusters: number of clusters used by KMeans.
    
    Returns:
      (accuracy, nmi, ari)
      - accuracy: float in [0,1], computed by finding the optimal one-to-one mapping
                  between cluster IDs and true labels via the Hungarian algorithm.
      - nmi: float in [0,1], normalized mutual information between gt and pred (ignoring label permutations).
      - ari: float in [−1,1], adjusted Rand index between gt and pred.
    """
    # Flatten ground-truth mask to shape (H*W,)
    gt_flat = gt_mask.flatten()
    pred = pred_labels.copy()

    # Determine unique true labels and remap them to indices [0..n_true-1]
    true_labels_unique = np.unique(gt_flat)
    n_true = len(true_labels_unique)
    label_to_index = {lab: idx for idx, lab in enumerate(true_labels_unique)}
    gt_indices = np.vectorize(label_to_index.get)(gt_flat)  # shape (H*W,)

    # Build confusion matrix: counts[i,j] = #pixels where pred==i and true==j
    counts = np.zeros((n_clusters, n_true), dtype=np.int64)
    for idx in range(pred.shape[0]):
        counts[pred[idx], gt_indices[idx]] += 1

    # Use Hungarian algorithm to find optimal one-to-one assignment between clusters and true labels
    cost_matrix = -counts  # maximize counts ↔ minimize negative counts
    row_ind, col_ind = linear_sum_assignment(cost_matrix)

    # Build mapping: cluster_id → true_label_value
    mapping = {cluster_id: true_labels_unique[true_idx] for cluster_id, true_idx in zip(row_ind, col_ind)}

    # Remap predicted cluster labels to the mapped true-label values
    pred_mapped = np.vectorize(lambda x: mapping[x])(pred)

    # Compute Accuracy
    acc = accuracy_score(gt_flat, pred_mapped)

    # Compute NMI and ARI on the original labels (they are permutation-invariant internally)
    nmi = normalized_mutual_info_score(gt_flat, pred)
    ari = adjusted_rand_score(gt_flat, pred)

    return acc, nmi, ari

def plot_metric_barchart(metrics_dict: dict):
    """
    Given a dictionary with keys ['Accuracy','NMI','ARI'] and values in [0,1],
    plot a bar chart showing these three metrics.
    """
    labels = list(metrics_dict.keys())
    values = [metrics_dict[k] for k in labels]

    plt.figure(figsize=(6, 4))
    bars = plt.bar(labels, values, alpha=0.8, edgecolor='black')
    plt.ylim(0, 1.0)
    plt.ylabel('Score')
    plt.title('Segmentation Metrics')
    for bar, v in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width() / 2, v + 0.02, f"{v:.2f}", ha='center')
    plt.tight_layout()
    plt.show()

def plot_intensity_histogram_with_centers(image: np.ndarray, centers: np.ndarray):
    """
    Plot histogram of pixel intensities (0..255) of the given image (flattened),
    and overlay dashed vertical lines at each cluster center.
    """
    plt.figure(figsize=(8, 4))
    # Histogram with 256 bins covering [0,255]
    plt.hist(image.flatten(), bins=256, range=(0, 255), alpha=0.7, edgecolor='black')

    # Sort centers so the lines appear left-to-right in order
    sorted_centers = np.sort(centers)
    for c in sorted_centers:
        plt.axvline(x=c, linestyle='--', linewidth=2, color='red')

    plt.title('Intensity Histogram with KMeans Cluster Centers')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

def show_results(orig: np.ndarray, clahe_img: np.ndarray,
                 raw_kmeans: np.ndarray, enhanced_norm: np.ndarray):
    """
    Display four grayscale images side by side with titles:
      [Original | CLAHE | Raw KMeans Segmentation | Enhanced (Normalized)].
    """
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    titles = ['Original', 'CLAHE', 'Raw KMeans', 'Enhanced (Norm)']
    imgs   = [orig, clahe_img, raw_kmeans, enhanced_norm]

    for ax, img, title in zip(axes, imgs, titles):
        ax.imshow(img, cmap='gray', vmin=0, vmax=255)
        ax.set_title(title)
        ax.axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # ============================================
    # 1. Load images (Grayscale and Ground Truth)
    # ============================================
    # NOTE: Replace these paths with your actual file locations.
    img_path     = '/Users/abubakr/Downloads/Spinal Cord Wedge Shape Plain.jpg'
    gt_mask_path = '/Users/abubakr/Downloads/Spinal Cord Wedge Shape Plain.jpg'
    n_clusters   = 8

    # Load the original grayscale image
    orig = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if orig is None:
        raise FileNotFoundError(f"Could not load image at '{img_path}'")

    # Load the ground-truth segmentation mask (single-channel, integer labels)
    gt_mask = cv2.imread(gt_mask_path, cv2.IMREAD_GRAYSCALE)
    if gt_mask is None:
        raise FileNotFoundError(f"Could not load ground-truth mask at '{gt_mask_path}'")

    if gt_mask.shape != orig.shape:
        raise ValueError("Ground-truth mask must have the same dimensions as the input image.")

    # ============================================
    # 2. Apply CLAHE for Local Contrast Enhancement
    # ============================================
    clahe_img = preprocess_image(orig)

    # ============================================
    # 3. Run KMeans on CLAHE-Enhanced Image
    # ============================================
    labels, centers, raw_kmeans, enhanced_norm = enhance_image_kmeans(clahe_img, n_clusters=n_clusters)

    # ============================================
    # 4. Compute Segmentation Metrics vs. Ground Truth
    # ============================================
    acc, nmi, ari = compute_segmentation_metrics(gt_mask, labels, n_clusters)
    print(f"Segmentation Metrics:\n  - Accuracy = {acc:.4f}\n  - NMI      = {nmi:.4f}\n  - ARI      = {ari:.4f}")

    # Plot bar chart summarizing those metrics
    metrics = {'Accuracy': acc, 'NMI': nmi, 'ARI': ari}
    plot_metric_barchart(metrics)

    # ============================================
    # 5. Show Image Results Side by Side
    # ============================================
    show_results(orig, clahe_img, raw_kmeans, enhanced_norm)

    # ============================================
    # 6. Plot Intensity Histogram with Cluster Centers
    # ============================================
    plot_intensity_histogram_with_centers(clahe_img, centers)






```
