```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import normalized_mutual_info_score, adjusted_rand_score, accuracy_score
from scipy.optimize import linear_sum_assignment
import os

def preprocess_image(image: np.ndarray) -> np.ndarray:
    clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
    return clahe.apply(image)

def enhance_image_kmeans(image: np.ndarray, n_clusters: int = 8):
    pixel_values = image.reshape(-1, 1).astype(np.float32)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(pixel_values)
    centers = kmeans.cluster_centers_.flatten().astype(np.float32)
    labels = kmeans.labels_
    segmented_raw = centers[labels].reshape(image.shape)
    return labels, centers, segmented_raw.astype(np.uint8)

def compute_segmentation_metrics(gt_mask: np.ndarray, pred_labels: np.ndarray, n_clusters: int):
    gt_flat = gt_mask.flatten()
    pred = pred_labels.copy()
    true_labels_unique = np.unique(gt_flat)
    n_true = len(true_labels_unique)
    label_to_index = {lab: idx for idx, lab in enumerate(true_labels_unique)}
    gt_indices = np.vectorize(label_to_index.get)(gt_flat)
    counts = np.zeros((n_clusters, n_true), dtype=np.int64)
    for idx in range(pred.shape[0]):
        counts[pred[idx], gt_indices[idx]] += 1
    cost_matrix = -counts
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    mapping = {cluster_id: true_labels_unique[true_idx] for cluster_id, true_idx in zip(row_ind, col_ind)}
    pred_mapped = np.vectorize(lambda x: mapping.get(x, 0))(pred)
    acc = accuracy_score(gt_flat, pred_mapped)
    nmi = normalized_mutual_info_score(gt_flat, pred)
    ari = adjusted_rand_score(gt_flat, pred)
    return acc, nmi, ari

def plot_comparative_barchart(metrics_dicts: dict, title: str, save_path: str = None):
    metric_names = ['Accuracy', 'NMI', 'ARI']
    method_names = list(metrics_dicts.keys())
    n_metrics = len(metric_names)
    n_methods = len(method_names)
    bar_width = 0.2
    x = np.arange(n_metrics)
    plt.figure(figsize=(8, 5))
    for i, method in enumerate(method_names):
        values = [metrics_dicts[method][m] for m in metric_names]
        plt.bar(x + i * bar_width, values, width=bar_width, label=method)
    plt.xticks(x + bar_width * (n_methods - 1) / 2, metric_names)
    plt.ylim(0, 1.05)
    plt.ylabel('Score')
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path)
        print(f"Saved bar chart to: {save_path}")
    plt.show()

def plot_intensity_histogram_with_centers(image: np.ndarray, centers: np.ndarray, title: str, save_path: str = None):
    plt.figure(figsize=(8, 4))
    plt.hist(image.flatten(), bins=256, range=(0, 255), alpha=0.7, edgecolor='black')
    sorted_centers = np.sort(centers)
    for c in sorted_centers:
        plt.axvline(x=c, linestyle='--', linewidth=2, color='red')
    plt.title(title)
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.tight_layout()
    if save_path:
        plt.savefig(save_path)
        print(f"Saved histogram to: {save_path}")
    plt.show()

def show_results(orig: np.ndarray, clahe_img: np.ndarray, raw_kmeans_img: np.ndarray, enh_kmeans_img: np.ndarray):
    fig, axes = plt.subplots(1, 4, figsize=(20, 4))
    titles = ['Original', 'CLAHE', 'Raw KMeans', 'Enhanced KMeans']
    imgs = [orig, clahe_img, raw_kmeans_img, enh_kmeans_img]
    for ax, img, title in zip(axes, imgs, titles):
        ax.imshow(img, cmap='gray', vmin=0, vmax=255)
        ax.set_title(title)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

def quantize_image_labels(image: np.ndarray, n_clusters: int) -> np.ndarray:
    bins = np.linspace(0, 256, n_clusters + 1, endpoint=True)
    labels = np.digitize(image.flatten(), bins) - 1
    labels = np.clip(labels, 0, n_clusters - 1)
    return labels

if __name__ == "__main__":
    # ==========================
    # PARAMETERS / PATHS
    # ==========================
    img_path = '/Users/abubakr/Downloads/WhatsApp Image 2025-06-02 at 10.52.55.jpeg'
    gt_mask_path = '/Users/abubakr/Downloads/WhatsApp Image 2025-06-02 at 10.52.55.jpeg'  # Replace if needed
    n_clusters = 8
    save_dir = '/Users/abubakr/Downloads'
    bar_chart_path = os.path.join(save_dir, 'segmentation_metric_comparison.png')
    hist_path = os.path.join(save_dir, 'clahe_histogram_with_kmeans_centers.png')

    # ==========================
    # LOAD IMAGES
    # ==========================
    orig = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if orig is None:
        raise FileNotFoundError(f"Could not load image at '{img_path}'")

    gt_mask = cv2.imread(gt_mask_path, cv2.IMREAD_GRAYSCALE)
    if gt_mask is None:
        raise FileNotFoundError(f"Could not load ground-truth mask at '{gt_mask_path}'")

    if gt_mask.shape != orig.shape:
        raise ValueError("Ground-truth mask must have the same dimensions as the input image.")

    # ==========================
    # APPLY CLAHE
    # ==========================
    clahe_img = preprocess_image(orig)

    # ==========================
    # KMEANS ON RAW AND ENHANCED
    # ==========================
    raw_labels, raw_centers, raw_kmeans_raw = enhance_image_kmeans(orig, n_clusters=n_clusters)
    enh_labels, enh_centers, enh_kmeans_raw = enhance_image_kmeans(clahe_img, n_clusters=n_clusters)

    # ==========================
    # BASELINE: QUANTIZE ORIGINAL INTO BINS
    # ==========================
    original_labels = quantize_image_labels(orig, n_clusters)

    # ==========================
    # COMPUTE METRICS
    # ==========================
    acc_orig, nmi_orig, ari_orig = compute_segmentation_metrics(gt_mask, original_labels, n_clusters)
    acc_raw, nmi_raw, ari_raw = compute_segmentation_metrics(gt_mask, raw_labels, n_clusters)
    acc_enh, nmi_enh, ari_enh = compute_segmentation_metrics(gt_mask, enh_labels, n_clusters)

    print(f"\nOriginal Image (Uniform Quantization) Metrics:\n"
          f"  - Accuracy = {acc_orig:.4f}\n"
          f"  - NMI      = {nmi_orig:.4f}\n"
          f"  - ARI      = {ari_orig:.4f}\n")

    print(f"\nRaw Image KMeans Metrics:\n"
          f"  - Accuracy = {acc_raw:.4f}\n"
          f"  - NMI      = {nmi_raw:.4f}\n"
          f"  - ARI      = {ari_raw:.4f}\n")

    print(f"\nEnhanced (CLAHE) Image KMeans Metrics:\n"
          f"  - Accuracy = {acc_enh:.4f}\n"
          f"  - NMI      = {nmi_enh:.4f}\n"
          f"  - ARI      = {ari_enh:.4f}\n")

    # ==========================
    # PLOT COMPARATIVE BARCHART
    # ==========================
    metrics_orig = {'Accuracy': acc_orig, 'NMI': nmi_orig, 'ARI': ari_orig}
    metrics_raw = {'Accuracy': acc_raw, 'NMI': nmi_raw, 'ARI': ari_raw}
    metrics_enh = {'Accuracy': acc_enh, 'NMI': nmi_enh, 'ARI': ari_enh}
    all_metrics = {
        'Original': metrics_orig,
        'Raw KMeans': metrics_raw,
        'Enhanced KMeans': metrics_enh
    }
    plot_comparative_barchart(all_metrics, title='Segmentation Metric Comparison', save_path=bar_chart_path)

    # ==========================
    # SHOW IMAGES SIDE-BY-SIDE
    # ==========================
    show_results(orig, clahe_img, raw_kmeans_raw, enh_kmeans_raw)

    # ==========================
    # PLOT HISTOGRAM OF ENHANCED IMAGE ONLY
    # ==========================
    plot_intensity_histogram_with_centers(
        clahe_img,
        enh_centers,
        title='CLAHE Image Histogram with Enhanced KMeans Centers',
        save_path=hist_path
    )
```
